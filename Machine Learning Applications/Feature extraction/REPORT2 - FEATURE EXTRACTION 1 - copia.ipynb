```python
print '%%% LOADING DATA %%%'
print '- Loading data.txt ...'
print '- Loading labels.txt ...'
# Load data
x = np.loadtxt( 'data.txt' )
y = np.loadtxt( 'labels.txt' )
```

```python
print '%%% PLOTTING THE DATA %%%'
# Plot the data
print '- Importing pyplot ...'
import matplotlib.pyplot as plt

%matplotlib inline

x_img = x[:,1].reshape(145, 145) # First band
plt.figure()
plt.imshow(x_img)
plt.title('X - Band 1')

y_img = y.reshape(145, 145) # Labels
plt.figure()
plt.imshow(y_img)
plt.title('Y - Band 1')

plt.show() # To show all figures
```

```python
print '%%% DATA PREPROCESSING %%%'
print '- Removing background class (class 0)'
# Remove Class 0
class_0 = np.nonzero(y != 0)
# Find matlab == where de numpy
x = x[class_0]
y = y[class_0]

# Checksum
print '- Checksum on X:', np.sum(np.sum(x,axis=0))
```

```python
print '%%% SPLIT DATA IN TRAINING-TEST PARTITIONS %%%'
# %%writefile nombre.py para guardar el .py
# Training 20% / Test 80%
print '- Training data (20%) ...'
print '- Test data (80%) ...'
N_data, N_dim = x.shape
N_train = round(0.2*N_data)
N_test = N_data-N_train

np.random.seed(0)
perm = np.random.permutation(N_data)
X_train = x[perm[:N_train],: ]
Y_train = y[perm[:N_train] ]
X_test = x[perm[N_train:],: ]
Y_test = y[perm[N_train:] ]

# Checksum
print '- Checksum on X_train:', np.sum(np.sum(X_train,axis=0))
```

```python
print '%%% Data normalization - Zero mean & unitary std %%%'
# Data normalization
mx = np.mean(X_train, axis=0, dtype=np.float64)
stdx = np.std(X_train, axis=0, dtype=np.float64)
print '- Normalizing train data ...'
print '- Normalizing test data ...'
X_train = np.divide((X_train-np.tile(mx, [N_train, 1])), np.tile(stdx, [N_train, 1]))
X_test = np.divide((X_test-np.tile(mx, [N_test, 1])), np.tile(stdx, [N_test, 1]))
```

```python
# Create output coding matrix Y
[nTrain, nBands] = X_train.shape
```

```python
# Linear FE + classifier
# Starting with PCA
from sklearn.decomposition import PCA

# class sklearn.decomposition.PCA(n_components=None, copy=True, whiten=False)
# n_components = 'mle' #Minka’s MLE algorithm
```

```python
# PCA range
nCompMax = nBands
nCompMin = 1
rang_PCA = np.arange(nCompMin, nCompMax+1, 5)

PCA_score_train = np.zeros((len(rang_PCA), 1))
PCA_score_test = np.zeros((len(rang_PCA), 1))
PCA_evr = np.zeros((len(rang_PCA), 1))

for i, nComp in enumerate(rang_PCA):
    #print nComp
    pca = PCA(n_components=nComp, whiten=False)
    pca.fit(X_train)
    # Score
    PCA_score_train[i] = pca.score(X_train)
    PCA_score_test[i] = pca.score(X_test)
    PCA_evr[i] = np.sum(pca.explained_variance_ratio_)
    
    
    
#print PCA_score_train
#print PCA_score_test
```
