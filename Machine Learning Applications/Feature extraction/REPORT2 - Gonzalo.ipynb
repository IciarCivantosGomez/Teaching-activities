```python
print '%%% IMPORTING PACKAGES %%%'
print '- Importing numpy ...'
print '- Importing neighbors from sklearn ...'
import numpy as np
from sklearn import neighbors
%connect_info

# import IPython.display as display
# display.YouTubeVideo('4p9MSJy761Y')
# !dir
# %load ipython_notebook_config.py
# %pdb para depurar
# %load http://matplotlib.org/mpl_examples/lines_bars_and_markers/marker_fillstyle_reference.py
```

```python
print '%%% LOADING DATA %%%'
print '- Loading data.txt ...'
print '- Loading labels.txt ...'
# Load data
x = np.loadtxt( 'data.txt' )
y = np.loadtxt( 'labels.txt' )
```

```python
print '%%% PLOTTING THE DATA %%%'
# Plot the data
print '- Importing pyplot ...'
import matplotlib.pyplot as plt

%matplotlib inline

x_img = x[:,1].reshape(145, 145) # First band
plt.figure()
plt.imshow(x_img)
plt.title('X - Band 1')

y_img = y.reshape(145, 145) # Labels
plt.figure()
plt.imshow(y_img)
plt.title('Y - Band 1')

plt.show() # To show all figures
```

```python
print '%%% DATA PREPROCESSING %%%'
print '- Removing background class (class 0)'
# Remove Class 0
class_0 = np.nonzero(y != 0)
# Find matlab == where de numpy
x = x[class_0]
y = y[class_0]

# Checksum
print '- Checksum on X:', np.sum(np.sum(x,axis=0))
```

```python
print '%%% SPLIT DATA IN TRAINING-TEST PARTITIONS %%%'
# %%writefile nombre.py para guardar el .py
# Training 20% / Test 80%
print '- Training data (20%) ...'
print '- Test data (80%) ...'
N_data, N_dim = x.shape
N_train = round(0.2*N_data)
N_test = N_data-N_train

np.random.seed(0)
perm = np.random.permutation(N_data)
X_train = x[perm[:N_train],: ]
Y_train = y[perm[:N_train] ]
X_test = x[perm[N_train:],: ]
Y_test = y[perm[N_train:] ]

# Checksum
print '- Checksum on X_train:', np.sum(np.sum(X_train,axis=0))
```

```python
print '%%% Data normalization - Zero mean & unitary std %%%'
# Data normalization
mx = np.mean(X_train, axis=0, dtype=np.float64)
stdx = np.std(X_train, axis=0, dtype=np.float64)
print '- Normalizing train data ...'
print '- Normalizing test data ...'
X_train = np.divide((X_train-np.tile(mx, [N_train, 1])), np.tile(stdx, [N_train, 1]))
X_test = np.divide((X_test-np.tile(mx, [N_test, 1])), np.tile(stdx, [N_test, 1]))
```

```python
############################################################
# PCA
############################################################
```

```python
# Create output coding matrix Y
[nTrain, nBands] = X_train.shape
```

```python
# Linear FE + classifier
# Starting with PCA
from sklearn.decomposition import PCA

# class sklearn.decomposition.PCA(n_components=None, copy=True, whiten=False)
# n_components = 'mle' #Minka’s MLE algorithm
```

```python
# PCA range
nCompMax = nBands
nCompMin = 1
rang_PCA = np.arange(nCompMin, nCompMax+1, 5)

PCA_score_train = np.zeros((len(rang_PCA), 1))
PCA_score_test = np.zeros((len(rang_PCA), 1))
PCA_evr = np.zeros((len(rang_PCA), 1))

for i, nComp in enumerate(rang_PCA):
    print nComp
    pca = PCA(n_components=nComp, whiten=False)
    pca.fit(X_train)
    # Score
    PCA_score_train[i] = pca.score(X_train)
    PCA_score_test[i] = pca.score(X_test)
    PCA_evr[i] = np.sum(pca.explained_variance_ratio_)
    
    
    
#print PCA_score_train
#print PCA_score_test
```

```python
plt.plot(rang_PCA, PCA_score_train, label='PCA_score_train')
plt.plot(rang_PCA, PCA_score_test, label='PCA_score_test')
#plt.show()
plt.xlabel('nComponents')
plt.ylabel('Log-likelihood (score)')
plt.legend()
#plt.savefig('./figures/KMEANS_K.eps')
#lt.savefig('./figures/KMEANS_K.pdf')
plt.savefig('./figures/LogPCA.eps', format='EPS')
plt.savefig('./figures/LogPCA.pdf', format='PDF')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show
#pint '- Y_score_test Max value Score=', np.max(Y_score_test), 'with K=', np.argmax(Y_score_test)+1
```

```python
# ESTO SON COSAS MIAS DEL DIA 11-3-2015...
# APLICAR UN CLASIFICADOR A VER QUE SALE...
# USAMOS EL RBF QUE IBA BIEN EN LA PRACTICA ANTERIOR
from sklearn import neighbors
from sklearn.grid_search import GridSearchCV

# PCA range
nCompMax = nBands
nCompMin = 1
nSkip = 5;
rang_PCA = np.arange(nCompMin, nCompMax+1, nSkip)

# KNN range
K_max = 10 # Neighbors
rang_K = np.arange(1, K_max+1)
KNN_CV_tuned_parameters = [{'n_neighbors': rang_K}]

# CV parameters
KNN_CV_nfold = 5
KNN_CV_score = 'precision'

print 'Band range for PCA:', rang_PCA
print 'Total bands:', nBands
print 'KNN neighbors range:', rang_K
```

```python
PCA_log_train = np.zeros((len(rang_PCA), 1))
PCA_log_test = np.zeros((len(rang_PCA), 1))
PCA_evr = np.zeros((len(rang_PCA), 1))
PCA_score_train = np.zeros((len(rang_PCA), 1))
PCA_score_test = np.zeros((len(rang_PCA), 1))
PCA_K_Best = np.zeros((len(rang_PCA), 1))


for i, nComp in enumerate(rang_PCA):
    # PCA PARA CADA nComp
    print '#Components', nComp
    pca = PCA(n_components=nComp, whiten=False)
    pca.fit(X_train)
    # Score
    PCA_log_train[i] = pca.score(X_train)
    PCA_log_test[i] = pca.score(X_test)
    PCA_evr[i] = np.sum(pca.explained_variance_ratio_)

    # Transform Training and Test data
    X_train_transformed = pca.transform(X_train)
    X_test_transformed = pca.transform(X_test)
    
    ###################################################
    # Cross Validation for KNN ranging 1-10 neighbors
    ###################################################
    print '--- KNN Classifier'
    # Blind Approach - CV done in Python
    
    KNN_CV_clf = GridSearchCV(neighbors.KNeighborsClassifier(1, weights='uniform'), KNN_CV_tuned_parameters, cv=KNN_CV_nfold, scoring=KNN_CV_score)
    KNN_CV_clf.fit(X_train_transformed, Y_train)
    PCA_K_Best[i] = KNN_CV_clf.best_estimator_.n_neighbors
    
    # Scores
    print '--- Scores...'
    PCA_score_train[i] = KNN_CV_clf.score(X_train_transformed, Y_train)
    PCA_score_test[i] = KNN_CV_clf.score(X_test_transformed, Y_test)
    
```

```python
#Best
print '%%% RESULTS KNN %%%'
print PCA_K_Best
print X_train_transformed.shape
```
