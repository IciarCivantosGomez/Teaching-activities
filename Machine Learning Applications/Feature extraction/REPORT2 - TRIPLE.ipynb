```python
print '%%% DATA PREPROCESSING %%%'
print '- Removing background class (class 0)'
# Remove Class 0
class_0 = np.nonzero(y != 0)
# Find matlab == where de numpy
x = x[class_0]
y = y[class_0]

# Checksum
print '- Checksum on X:', np.sum(np.sum(x,axis=0))
```

```python
print '%%% SPLIT DATA IN TRAINING-TEST PARTITIONS %%%'
# %%writefile nombre.py para guardar el .py
# Training 20% / Test 80%
print '- Training data (20%) ...'
print '- Test data (80%) ...'
N_data, N_dim = x.shape
N_train = round(0.2*N_data)
N_test = N_data-N_train

np.random.seed(0)
perm = np.random.permutation(N_data)
X_train = x[perm[:N_train],: ]
Y_train = y[perm[:N_train] ]
X_test = x[perm[N_train:],: ]
Y_test = y[perm[N_train:] ]

# Checksum
print '- Checksum on X_train:', np.sum(np.sum(X_train,axis=0))
```

```python
print '%%% Data normalization - Zero mean & unitary std %%%'
# Data normalization
mx = np.mean(X_train, axis=0, dtype=np.float64)
stdx = np.std(X_train, axis=0, dtype=np.float64)
print '- Normalizing train data ...'
print '- Normalizing test data ...'
X_train = np.divide((X_train-np.tile(mx, [N_train, 1])), np.tile(stdx, [N_train, 1]))
X_test = np.divide((X_test-np.tile(mx, [N_test, 1])), np.tile(stdx, [N_test, 1]))
```

```python
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # 
#My implementation (includes sigma estimation)
def rbf_kernel_sig(X1, X2, sig=0):
    size1=X1.shape[0];
    size2=X2.shape[0];
    if X1.ndim==1:
        X1=X1[:,np.newaxis]
        X2=X2[:,np.newaxis]
    G=(X1* X1).sum(axis=1)
    H=(X2* X2).sum(axis=1)
    Q = np.tile(G, [size2,1]).T
    R = np.tile(H, [size1,1])
    KK=np.dot(X1,X2.T)
    dist=(Q + R - 2*KK)
    if sig==0:  # Then, we estimate its value
        aux=dist-np.tril(dist)
        aux=aux.reshape(size1**2,1)
        sig=np.sqrt(0.5*np.mean(aux[np.where(aux>0)]))             
    K = np.exp(-dist/sig**2);
    return K, sig
# 
# Scikit learn implementation
# from sklearn.metrics.pairwise import rbf_kernel
# K=rbf_kernel(X1,X2,gamma=1/(sig_x**2))
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # 
```

```python
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.preprocessing import KernelCenterer
from scipy.linalg import eigh
```

```python
# TRAINING KERNEL
print 'TRAINING KERNEL'
K_vanessa,sig_vanessa = rbf_kernel_sig(X_train, X_train, sig=0)

# Center the kernel matrix
kern_cent = KernelCenterer()
K_v = kern_cent.fit_transform(K_vanessa)
print 'Center kernel...'
# Centering the symmetric NxN kernel matrix.
N = K_vanessa.shape[0]
one_n = np.ones((N,N)) / N
K_norm = K_vanessa - one_n.dot(K_vanessa) - K_vanessa.dot(one_n) + one_n.dot(K_vanessa).dot(one_n)

####################################################################################################
print 'TEST KERNEL'
# TEST KERNEL
K_vanessa_test,sig_vanessa_test = rbf_kernel_sig(X_test, X_test, sig=0)

# Center the kernel matrix
kern_cent_test = KernelCenterer()
K_v_test = kern_cent_test.fit_transform(K_vanessa_test)
print 'Center kernel...'
# Centering the symmetric NxN kernel matrix.
N = K_vanessa_test.shape[0]
one_n = np.ones((N,N)) / N
K_norm_test = K_vanessa_test - one_n.dot(K_vanessa_test) - K_vanessa_test.dot(one_n) + one_n.dot(K_vanessa_test).dot(one_n)
```

```python
def kernel_eigen(K_in):
    eigvals, eigvecs = eigh(K_in)
    
    return eigvals, eigvecs

def devolver_eigen_kernel(eigvals, eigvecs, numComponentes):
    # Obtaining eigenvalues in descending order with corresponding 
    # eigenvectors from the symmetric matrix.
    #eigvals, eigvecs = eigh(K_in)

    # Obtaining the i eigenvectors (alphas) that corresponds to the i highest eigenvalues (lambdas).
    # alpha - eigenvectors
    alphas = np.column_stack((eigvecs[:,-i] for i in range(1,numComponentes+1)))
    # lambdas - eigenvalues
    lambdas = [eigvals[-i] for i in range(1,numComponentes+1)]
    
    return alphas, lambdas

print 'eig train'
eigvals_train, eigvecs_train = kernel_eigen(K_norm)
print 'eig test'
eigvals_test, eigvecs_test = kernel_eigen(K_norm_test)
```

```python
from sklearn import svm, grid_search
# .. classification ..
rang_C = np.logspace(0, 3, 10)
rang_Gamma = np.logspace(-5,-2, 5)
parameters_linear_SVM_pca = {'C':rang_C, 'gamma':rang_Gamma}
nFolds = 5
# PCA range
[nTrain, nBands] = X_train.shape
nCompMax = nBands
nCompMin = 1
nSkip = 10;
rang_PCA = np.arange(nCompMin, nCompMax+1, nSkip)
# Scores
Kernel_score_train = np.zeros((len(rang_PCA), 1))
Kernel_score_test = np.zeros((len(rang_PCA), 1))

for i, nComp in enumerate(rang_PCA):

    print  'i: ', i, 'nComp: ', nComp
    print '--- CV...'
    svr_linear_SVM_pca = svm.SVC(kernel='rbf')
    clf = grid_search.GridSearchCV(svr_linear_SVM_pca, parameters_linear_SVM_pca, cv=nFolds)
    
    # GET alphas and lambdas
    print '--- train: Getting alphas, lambdas...'
    alphas_train, lambdas_train = devolver_eigen_kernel(eigvals_train, eigvecs_train, nComp)
    print '--- test: Getting alphas, lambdas...'
    alphas_test, lambdas_test = devolver_eigen_kernel(eigvals_test, eigvecs_test, nComp)
    
    '--- fitting...'
    clf.fit(alphas_train,Y_train)
    print '--- score...'
    Kernel_score_train[i] = clf.score(alphas_train, Y_train)
    Kernel_score_test[i] = clf.score(alphas_test, Y_test)
    
```

```python
plt.plot(rang_PCA, Kernel_score_train, label='Kernel score train')
plt.plot(rang_PCA, Kernel_score_test, label='Kernel score test')
#plt.show()
plt.xlabel('nComponents')
plt.ylabel('Kernel CV (score)')
plt.legend()
#plt.savefig('./figures/KMEANS_K.eps')
#lt.savefig('./figures/KMEANS_K.pdf')
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.savefig('./figures/Kernel_SVM_CV.eps', format='EPS')
plt.savefig('./figures/Kernel_SVM_CV.pdf', format='PDF')
plt.savefig('./figures/Kernel_SVM_CV.png', format='PNG')
```
